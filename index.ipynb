<script>
(function() {
  function initBanner() {
    var hero = document.querySelector('.resume-hero');
    var canvas = document.getElementById('resume-banner');
    if (!hero || !canvas) return;

    var ctx = canvas.getContext('2d');
    var dpr = window.devicePixelRatio || 1;
    var nodes = [];
    var edges = [];
    var layers = [3, 6, 2];
    var breathDuration = 5.5; // seconds for one calm fade-in / fade-out breath
    var breathShape = 0.5;  // <1 = stay visible longer, fade faster near ends
    var lastBreathIndex = -1;

    function resize() {
      var w = hero.clientWidth;
      var h = hero.clientHeight;
      canvas.width = Math.floor(w * dpr);
      canvas.height = Math.floor(h * dpr);
      ctx.setTransform(1, 0, 0, 1, 0, 0);
      ctx.scale(dpr, dpr);

      nodes = [];
      edges = [];

      // Keep the network clustered in the upper-right corner of the hero box.
      var region = {
        xMin: w * 0.59,
        xMax: w * 0.96,
        yMin: h * 0.01,
        yMax: h * 0.47
      };

      layers.forEach(function(count, li) {
        var x = region.xMin + (region.xMax - region.xMin) * (li / (layers.length - 1));
        for (var i = 0; i < count; i++) {
          var y = region.yMin + (region.yMax - region.yMin) * ((i + 1) / (count + 1));
          nodes.push({
            x: x,
            baseY: y,
            layer: li,
            r: 3 + Math.random() * 1.5,
            speed: 0.4 + Math.random() * 0.6,
            phase: Math.random() * Math.PI * 2,
            amp: 1.5 + Math.random() * 2
          });
        }
      });

      generateEdges();
    }

    function generateEdges() {
      edges = [];
      var hasInput = {};

      // Input → hidden.
      for (var i = 0; i < nodes.length; i++) {
        var a = nodes[i];
        if (a.layer !== 0) continue;
        for (var j = 0; j < nodes.length; j++) {
          var b = nodes[j];
          if (b.layer === a.layer + 1 && Math.random() < 0.65) {
            edges.push({ i: i, j: j, layer: a.layer, phase: Math.random() * 0.8 });
            hasInput[j] = true;
          }
        }
      }

      // Hidden → output only for hidden nodes that received input.
      for (var i = 0; i < nodes.length; i++) {
        var a = nodes[i];
        if (a.layer !== 1 || !hasInput[i]) continue;
        for (var j = 0; j < nodes.length; j++) {
          var b = nodes[j];
          if (b.layer === a.layer + 1 && Math.random() < 0.65) {
            edges.push({ i: i, j: j, layer: a.layer, phase: Math.random() * 0.8 });
          }
        }
      }
    }

    function draw() {
      var w = hero.clientWidth;
      var h = hero.clientHeight;
      var t = performance.now() / 1000;
      var breathIndex = Math.floor(t / breathDuration);

      if (breathIndex !== lastBreathIndex) {
        generateEdges();
        lastBreathIndex = breathIndex;
      }

      var breathProgress = (t % breathDuration) / breathDuration;
      var breath = Math.pow(Math.sin(breathProgress * Math.PI), breathShape);

      ctx.clearRect(0, 0, w, h);

      var live = nodes.map(function(n) {
        return {
          x: n.x,
          y: n.baseY + Math.sin(t * n.speed + n.phase) * n.amp,
          r: n.r
        };
      });

      // Animate edge opacity like a signal activating across the network.
      // The pulse is anchored to the breath cycle so edges fade out cleanly
      // before the set is re-initialized at the start of the next breath.
      edges.forEach(function(e) {
        var a = live[e.i];
        var b = live[e.j];
        var pulse = 0.5 + 0.5 * Math.sin(breathProgress * Math.PI * 2 * 2 - e.layer * 1.6 + e.phase);
        var alpha = (0.25 + 0.5 * pulse) * breath;
        ctx.strokeStyle = 'rgba(233, 84, 32, ' + alpha + ')';
        ctx.lineWidth = (1.2 + pulse * 0.8) * breath;
        ctx.beginPath();
        ctx.moveTo(a.x, a.y);
        ctx.lineTo(b.x, b.y);
        ctx.stroke();
      });

      live.forEach(function(n) {
        ctx.beginPath();
        ctx.arc(n.x, n.y, n.r, 0, Math.PI * 2);
        ctx.fillStyle = 'rgba(233, 84, 32, 0.65)';
        ctx.fill();
      });

      requestAnimationFrame(draw);
    }

    resize();
    window.addEventListener('resize', resize);
    draw();
  }

  if (document.readyState === 'loading') {
    document.addEventListener('DOMContentLoaded', initBanner);
  } else {
    initBanner();
  }
})();
</script>
