# # End-to-End Integration Demo

Walking through the complete demo.

## The demo flow

1. **Deploy Azure stack**: `make az-up` — Container Apps, PG, Blob, Redis, Functions, KV, ACR
2. **Train a model**: run `train.py` against Azure-hosted MLflow, register model + prompt
3. **Publish batch tasks**: producer writes 30 documents to Redis
4. **Observe KEDA**: workers scale from 0, process tasks, log to Application Insights
5. **Verify idempotency**: publish duplicate tasks — idempotency guard skips them
6. **Simulate crash**: `docker compose restart redis` / restart Redis Container App —
   reconciliation producer recovers lost queue state
7. **Online inference**: call Azure Functions endpoint with a single document
8. **Dashboard**: walk through Application Insights dashboards, KQL queries, alerts
9. **Tear down**: `make az-down` — everything gone, bill stops
