# # The Idempotency Guard

Implementing exactly-once side effects on top of at-least-once delivery.

## Topics

- The problem: Redis + Celery = at-least-once. Your side effects need exactly-once.
- The mechanism: `SETNX` (SET-if-Not-eXists) with a TTL-aligned expiry
- The `idempotency_guard` context manager — full implementation
- What happens on duplicate delivery, worker crash, Redis restart
- Why 7-day TTL: matching idempotency window to result expiry window
- Testing: simulate duplicate delivery, verify no double side effects
- When `SETNX` is NOT enough (external API idempotency, idempotency-key pass-through)

## Prerequisites

Celery + Redis setup from [06-celery-redis](06-celery-redis.ipynb).
