# # MLflow 3 — Tracking, Registry, and Prompts

Deep dive into the MLflow 3 server running in Docker Compose.

## Topics

- Experiment tracking: logging parameters, metrics, artifacts, tags, and lineage
- Model registry: immutable versions, aliases (candidate/staging/champion), promotion workflow
- Prompt registry: versioned prompts with immutable references
- LLM tracing: tracing GenAI calls through MLflow
- Querying runs, comparing experiments, and downloading artifacts programmatically

## Prerequisites

Docker Compose stack running from [02-local-stack](02-local-stack.ipynb).
