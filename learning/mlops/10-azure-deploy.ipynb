# # Deploying and Tearing Down

Provisioning the Azure stack and tearing it down when done.

## Topics

- `az deployment group create` with Bicep entry point
- Post-deploy: create MinIO-equivalent blob container, set up MLflow artifact proxy
- Smoke tests after deploy: MLflow health, Redis ping, PG connection, OpenAI call
- Setting MLFLOW_TRACKING_URI to the Azure-hosted MLflow
- Tear-down: `az group delete` or `az deployment group delete` + individual resource cleanup
- Cost-optimization strategies: stop PG when not in use, use consumption tiers
- Scripting deploy and teardown with Makefile targets: `make az-up`, `make az-down`

## Prerequisites

- Bicep templates from [09-azure-iac](09-azure-iac.ipynb)
- All local components (02–08) tested and working
