# # Celery + Redis — Task Queue Fundamentals

Setting up Celery with Redis as broker and result backend.

## Topics

- Celery architecture: broker, result backend, workers, producers
- Redis as broker: list-based queues, at-least-once delivery
- Redis as result backend: 7-day TTL, key patterns, memory sizing
- Configuration: `acks_late`, `task_reject_on_worker_lost`, `prefetch_multiplier`
- Task contracts: version, release, model/prompt versions, idempotency key
- The `process_document` task signature and why each field matters
- Running workers and publishing tasks from Python

## Prerequisites

Docker Compose stack from [02-local-stack](02-local-stack.ipynb).
