# # Bicep — Infrastructure as Code

Translating the Docker Compose stack to Azure Bicep templates.

## Topics

- Bicep syntax primer: resources, parameters, modules, outputs
- Container Apps Environment + Container App (MLflow)
- Azure Database for PostgreSQL Flexible Server (B1ms, burstable)
- Azure Blob Storage account + container for MLflow artifacts
- Key Vault for secrets (PG password, Redis password, OpenAI key)
- Azure Container Registry for immutable container images
- Azure Functions app (consumption plan)
- Network: VNet, subnets, private endpoints where needed
- Parameterizing for environment (dev/prod) and cost controls

## Prerequisites

Azure CLI (`az`) installed and authenticated. An active Azure subscription.
