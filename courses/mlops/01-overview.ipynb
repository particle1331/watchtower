# ML Platform on Azure — Learning Journey

A step-by-step build of a lightweight MLOps platform for personal learning and demo.

## What we're building

A portable ML platform centered on **MLflow 3** and **Celery + Redis** task queues,
deployed to **Azure Container Apps**:

- MLflow 3 on Container Apps for experiment tracking, model registry, prompts, and LLM tracing
- PostgreSQL + Blob Storage as MLflow's backend
- Celery + Redis + KEDA for batch inference that scales to zero
- Azure Functions for online inference
- Azure OpenAI for LLM calls
- Application Insights + Azure Monitor for observability
- Bicep for infrastructure-as-code

## Local vs Azure

**Local** (steps 02–08): Docker Compose with MLflow, PostgreSQL, Redis
(single container), and MinIO (S3-compatible artifact store). Fast iteration —
change code, `docker compose restart`, test.

**Azure** (steps 09–13): The same stack mapped to managed services:

| Local | Azure |
|---|---|
| Docker Compose | Bicep templates |
| Postgres container | Azure Database for PostgreSQL (B1ms) |
| MinIO | Azure Blob Storage |
| Redis container | Redis on Container Apps |
| MLflow server container | MLflow on Container Apps |
| Celery workers in Compose | Celery on Container Apps + KEDA |
| — | Azure Functions |
| — | Application Insights + Azure Monitor |
| — | Key Vault + ACR |
