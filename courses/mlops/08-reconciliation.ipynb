# # Reconciliation Producer — Crash Recovery

Recovering lost queue state after Redis restarts with Celery Beat.

## Topics

- What Redis restart loses: queued-but-unacknowledged messages
- What survives: in-flight tasks (they reconnect), completed results (in Redis or logs)
- The Beat reconciliation schedule: scan unprocessed documents, diff against Redis result keys,
  republish missing work with the same idempotency key
- Why reconciliation is safe: idempotency keys absorb any duplicates
- Testing: `docker compose restart redis`, observe Beat recovery
- Window alignment: reconciliation interval == result TTL == idempotency TTL

## Prerequisites

- Celery + Redis from [06-celery-redis](06-celery-redis.ipynb)
- Idempotency guard from [07-idempotency](07-idempotency.ipynb)
