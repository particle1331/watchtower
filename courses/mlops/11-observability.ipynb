# # Observability — Dashboards, Alerts, and KQL

Monitoring the Azure ML platform with Application Insights and Azure Monitor.

## Topics

- Application Insights: structured telemetry, trace context, custom dimensions
- Logging from Celery workers: correlation ids, task metadata, token usage, latency
- Azure Monitor: creating dashboards from platform metrics
- KQL (Kusto Query Language): querying Log Analytics for traces and custom events
- Alerts: error rate, queue depth (LLEN), Redis health, PG CPU, Azure OpenAI throttling
- Sampling: cost-conscious telemetry without losing critical events
- Dashboard walkthrough: one pane covering MLflow, Redis, PG, workers, and OpenAI

## Prerequisites

Azure stack deployed from [10-azure-deploy](10-azure-deploy.ipynb).
