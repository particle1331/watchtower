# ML Platform on Azure

A step-by-step build of a lightweight MLOps platform for personal learning and demo.

## What we're building

A portable ML platform centered on **MLflow 3** and **Celery + Redis** task queues,
deployed to **Azure Container Apps**:

- MLflow 3 on Container Apps for experiment tracking, model registry, prompts, and LLM tracing
- PostgreSQL + Blob Storage as MLflow's backend
- Celery + Redis + KEDA for batch inference that scales to zero
- Azure Functions for online inference
- Azure OpenAI for LLM calls
- Application Insights + Azure Monitor for observability
- Bicep for infrastructure-as-code

## Learning path

### Local Stack (02–08)

| Step | Topic | Key Topics |
|---|---|---|
| 02 | Docker Compose Local Stack | MLflow, PostgreSQL, Redis, MinIO in one command |
| 03 | MLflow 3: Tracking, Registry, Prompts | Experiment tracking, model registry, prompt registry, LLM tracing |
| 04 | Data Validation with Pandera | Data contracts, schema evolution, validation reporting |
| 05 | ML Pipeline: Training & Packaging | Scikit-learn pipeline, MLflow packaging, training-serving skew |
| 06 | Celery + Redis: Task Queue | Broker, result backend, at-least-once delivery, task contracts |
| 07 | The Idempotency Guard | SETNX, exactly-once side effects, duplicate detection |
| 08 | Reconciliation Producer | Crash recovery, Celery Beat, lost queue state |

### Azure Deployment (09–13)

| Step | Topic | Key Topics |
|---|---|---|
| 09 | Bicep: Infrastructure as Code | Container Apps, PG, Blob, KV, ACR, network |
| 10 | Deploying and Tearing Down | az deployment, smoke tests, Makefile targets, cost optimization |

| 12 | KEDA Scaling | Redis list scaler, scale-to-zero, capacity planning |
| 13 | End-to-End Integration Demo | Full demo walkthrough, crash simulation, recording |

- Docker and Docker Compose
- Python 3.10+ with basic ML knowledge
- Azure CLI with an active subscription (~$0.50 per demo session)

## Budget

~$0.50 per demo session with tear-down discipline (~$10–20 total for the whole journey).
