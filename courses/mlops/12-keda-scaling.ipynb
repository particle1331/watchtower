Auto-scaling Celery workers based on Redis queue depth.

## Topics

- KEDA architecture: ScaledObject, triggers, scaling rules
- The Redis list scaler: `listLength` trigger on versioned Redis keys
- Configuration: min=0 replicas, max derived from Azure OpenAI quota
- Watching scaling in action: publish N tasks, observe Container Apps scale from 0
- Scale-to-zero: what happens to in-flight tasks when KEDA scales to zero?
- Safe shutdown: `SIGTERM` handling, `acks_late`, warm shutdown with Container Apps
- Prefetch multiplier = 1: one task per worker, scaling is predictable
- Capacity planning: p95 tokens per doc x replica count x concurrency <= RPM quota

## Prerequisites

- Azure stack from [10-azure-deploy](10-azure-deploy.ipynb)
- Celery + Redis code from [06-celery-redis](06-celery-redis.ipynb) and [07-idempotency](07-idempotency.ipynb)
